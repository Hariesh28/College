```python
n = 4

board = [[0] * n for _ in range(n)]
board
```

```python
def is_valid(num: int, row: int, col: int, board: list[list[int]]) -> bool:

    # Check Row
    if num in board[row]:
        return False

    # Check column
    for i in range(4):
        if board[i][col] == num:
            return False

    start_row = (row // 2) * 2
    start_col = (col // 2) * 2

    # Check 2x2 box
    for i in range(start_row, start_row+2):
        for j in range(start_col, start_col+2):
            if board[i][j] == num:
                return False

    return True
```

```python
def find_empty(board):

    for i in range(4):
        for j in range(4):
            if board[i][j] == 0:
                return i, j

    return None
```

```python
def print_board(board):
    for row in board:
        print(row)
    print()
```

```python
def solve(board: list[list[int]], solutions: list[list[int]]):

    empty = find_empty(board)
    if empty is None:
        solutions.append([row[:] for row in board])
        return

    row, col = empty

    for i in range(1, 5):
        if is_valid(i, row, col, board):
            board[row][col] = i
            solve(board, solutions)
            board[row][col] = 0
```

```python
sol = []
solve(board, sol)
```

```python
for i in sol:
    print_board(i)
    print()
```
